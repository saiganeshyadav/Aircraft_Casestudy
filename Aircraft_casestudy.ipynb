```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel('Data Science Case study Vindiata.xlsx',sheet_name=['AC characteristics','Operations','City pairs'])
```

```python
AC_characteristics_df = data.get('AC characteristics')
Operations_df = data.get('Operations')
City_pairs_df = data.get('City pairs')
```

```python
AC_characteristics_df.head()
```

```python
AC_characteristics_df.drop('Aircraft Costs',axis=1,inplace=True)
AC_characteristics_df.drop(AC_characteristics_df.iloc[:,5:],axis=1,inplace=True)
AC_characteristics_df.drop(index=AC_characteristics_df.index[:2],inplace=True)
AC_characteristics_df.columns = AC_characteristics_df.iloc[0,:].values
AC_characteristics_df.drop([2],axis=0,inplace=True)
AC_characteristics_df.reset_index(drop=True,inplace=True)
```

```python
AC_characteristics_df
```

```python
Operations_df
```

```python
Operations_df.drop('Airline A - hours flown',axis=1,inplace=True)
Operations_df.drop(Operations_df.iloc[:,14:],axis=1,inplace=True)
Operations_df.drop(index=Operations_df.index[:2],inplace=True)
Operations_df.columns = Operations_df.iloc[0,:].values
Operations_df.drop([2],axis=0,inplace=True)
Operations_df.reset_index(drop=True,inplace=True)
```

```python
Operations_df
```

```python
City_pairs_df
```

```python
City_pairs_df.drop(City_pairs_df.columns[0],axis=1,inplace=True)
City_pairs_df.drop(City_pairs_df.iloc[:,4:],axis=1,inplace=True)
City_pairs_df.drop(index=City_pairs_df.index[:2],inplace=True)
City_pairs_df.columns = City_pairs_df.iloc[0,:].values
City_pairs_df.drop([2],axis=0,inplace=True)
City_pairs_df.reset_index(drop=True,inplace=True)
```

```python
City_pairs_df
```

```python
dic = {}
total_hours=[]
for i in range(len(AC_characteristics_df)):
    df = Operations_df.groupby('Aircraft Type').sum()
    hours = df.iloc[i,1:].values.sum()
    cost = AC_characteristics_df[AC_characteristics_df['Aircraft Type'] == df.index[i]]['Costs per flight hour'].values[0]
    total = hours * cost
    total_hours.append(hours)
    dic[df.index[i]] = total
```

```python
total_cost_df = pd.DataFrame()
total_cost_df['Aircraft Type'] = dic.keys()
total_cost_df['Total_hours'] = total_hours
total_cost_df['Total_Amount(in $)'] = dic.values()
total_cost_df
```

**These are the Total Amounts(in $) of each Aircraft Type for the year**

```python
total_cost_df = total_cost_df.set_index('Aircraft Type')
total_cost_df = total_cost_df.reindex(index=AC_characteristics_df['Aircraft Type'])
total_cost_df = total_cost_df.reset_index()
```

```python
total_cost_df
```

```python
cost_per_seat_per_km_flown = (total_cost_df['Total_Amount(in $)'])/(total_cost_df['Total_hours'] * AC_characteristics_df['Ave. Speed (km/h)'] * AC_characteristics_df['Number of Seats'])
```

```python
total_cost_df['cost_per_seat_per_km_flown(in $)'] = cost_per_seat_per_km_flown
total_cost_df
```

### From above dataframe we can see that A330 Aircraft Type is having the lowest cost per seat per km flown with cost of 0.033333 $

```python
air_type=[]
no_trips=[]
cost_per_trip=[]
ori_des_pair=[]
for i in range(len(City_pairs_df)):
    d = AC_characteristics_df[AC_characteristics_df['Range (Km)'] >= City_pairs_df.loc[i,'Distance (km)']]
    for j in d.index:
        ori_des_pair.append(City_pairs_df.loc[i,'Origin City']+'-'+City_pairs_df.loc[i,'Desitnation City'])
        air_type.append(d.loc[j,'Aircraft Type'])
        no_trips.append(int(np.ceil(City_pairs_df.loc[i,'Pass. Demand \n(per day)'] / d.loc[j,'Number of Seats'])))
        cost_per_trip.append(d.loc[j,'Number of Seats'] * total_cost_df.loc[j,'cost_per_seat_per_km_flown(in $)'] * City_pairs_df.loc[i,'Distance (km)'])
        
```

```python
final_df = pd.DataFrame()
final_df['ori_des_pair'] = ori_des_pair
final_df['air_type'] = air_type
final_df['NO. of trips will take'] = no_trips
final_df['cost_per_trip'] = cost_per_trip
final_df['Total_cost(in $)'] = final_df['NO. of trips will take'] * final_df['cost_per_trip']
```
